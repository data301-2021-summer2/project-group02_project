### Graphs of State Births

```python
state_births_month, df_state_year, df_state_births=project_functions.load_and_process("allBirthData.csv")
# 
x=df_state_births.index
y=df_state_births["births"]

plt.figure(figsize=(10,8))
sns.barplot(x,y)
plt.xticks(fontsize=8)
plt.xlabel("State")
plt.ylabel("Births")
plt.title("State-births")
plt.savefig('Bar Graph of All State-birth.png')
plt.show()
```

### Scatterplot of 51 states divided into 4 graphs. Yearly states births are shown in differnt points.

```python
# scatterplot
df_state_births=df_state_births.sort_values(by="births",ascending=False)


series_plot1=df_state_births.index[0:13]
plt.figure(figsize=(10,8))
plt.xlabel("Year")
plt.ylabel("Births")
plt.title("13 states' births")
sns.scatterplot(data=df_state_year.loc[:,series_plot1])
plt.savefig('Scatter-Graph-1,13-States-birth.png')
plt.show()

series_plot2=df_state_births.index[13:26]
plt.figure(figsize=(10,8))
plt.xlabel("Year")
plt.ylabel("Births")
plt.title("13 states' births")
sns.scatterplot(data=df_state_year.loc[:,series_plot2])
plt.savefig('Scatter-Graph-2,13-States-birth.png')
plt.show()

series_plot3=df_state_births.index[26:39]
plt.figure(figsize=(10,8))
sns.scatterplot(data=df_state_year.loc[:,series_plot3])
plt.xlabel("Year")
plt.ylabel("Births")
plt.title("13 states' births")
plt.savefig('Scatter-Graph-3,13-States-birth.png')
plt.show()

series_plot4=df_state_births.index[39:]
plt.figure(figsize=(10,8))
plt.xlabel("Year")
plt.ylabel("Births")
plt.title("12 states' births")
sns.scatterplot(data=df_state_year.loc[:,series_plot4])
plt.savefig('Scatter-Graph-4,12-States-birth.png')
plt.show()
```
